# Write up

For this project, my goal was to find the most undervalued players in the NBA for a team to sign. I set up a multi-pronged analytical approach using data-driven methods to determine the best players at the best value. My goal is to optimize player ability and value to build a sustainable, winning NBA team. First, I approached this problem by finding a relevant statistic that could be applied to all players as a common basis for their skills. I created a stat called 'player efficiency rating' for my dataset. This statistic combined a playerâ€™s positive achievements with their negative achievements, leading to the correct weighting of each for their perceived worth. There also needed to be a way to more highly value players who had repeatedly and consistently proven themselves. This would ensure that a player who shot 50% on four shots was not measured the same as a player who shot 50% on 1,000 shots. Hence, I made it so players had to play a certain number of minutes. Then, I made it so that each player was placed into two quantiles: above median or below median, with below median shooting penalized. There was also a minutes requirement that was included in the dataset, where players who failed to meet the minutes requirement were not included, as their numbers were based on too little play. This helped eliminate outliers that would skew the set when there was little data on the player. I also adjusted the weight of each stat in the positive achievements category; for example, offensive rebounds were more valuable than defensive rebounds, and turnovers were weighted the highest because they cause a full loss of possession. I then was able to create the player efficiency metric, which was then plotted against salary and put into clusters to see which section each player would fall into. I ran an elbow curve test to determine the correct amount of clusters to use, which came out to 3, as that is where there was an elbow in the  graph, as well as a strong silhouette score. The elbow curve helped identify the most undervalued players. These plots and graphs can be seen in the lab clustering tab in the notebook, which provides a full visual representation of the analysis. They can also be seen below. Now that I had the optimal number of clusters, I could then plot to have 3 clusters to find the most undervalued section, which came out to be the red section with around 90 or so players. The players seen in the red dots in the chart are high efficiency and lower cost. These are the target players. From there, I was able to print out those players' names to find the most undervalued players for my team to attempt to sign. Given the player efficiency metric, those that ranked the highest efiiciency per dollar were Jamal Shead, Craig Porter Jr., and Isaiah Collier. These would be the most desirable on the developed ranking system, and the next three players to target would be Dru Smith, Tyler Kolek, and Vince Williams Jr. On the other end of the spectrum, the players with low effciency per dollar are OG Anunoby, Jaren Jackson Jr., Paul George, and Jordan Poole. These would be players to shy away from in recruiting. In the end, using this methodology will lead to a team that overperforms their value with the potential to go all the way to an NBA Championship. The data to support this decision is there and can not be overlooked and we generated the neighborhood of players to choose from to lead to success.
