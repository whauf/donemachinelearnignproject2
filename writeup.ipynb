# Write up

For this project my goal was to find the most undervalued players in the NBA for a team to sign. I approached this problem by first trying to find a possible statistic that was able to carry over to all players to be a common basis for their skill. I created a stat that I called player efficiency rating for my data set, which combined a player’s positive achievements against their negative, correctly weighting each for its perceived worth. There also needed to be a way to value players who had proven themselves more so that a player who shot 50% on 4 shots wasn’t measured the same as a player who shot 50% off 1000 shots. I first made it so players had to play a certain number of minutes. Then, I made it so that each player was placed into two quantiles, shooting above median, or below, which below was penalized. There was also a minutes requirement that was put into the dataset where players who failed to meet the minites requirement were not put into the dataset as their numbers were based on too few minutes. This helped to eliminate outliars that would skew the set without significant data on the player. I also adjusted the weight of each stat that was put into the positive achievements category, for example the offensive rebounds were more valuable than defensive rebounds, and turnovers were weighted the highest as that causes a full loss of possession. I then was able to create the player efficiency metric which was then able to be plotted against salary and put into clusters to see what section each player would fall into. I ran an elbow curve test to determine the correct amount of clusters to use which came out to 3, as that is where there was an elbow in the attached graph as well as a strong silhouette score. Now that I had the most optimal amount of clusters I could then plot to have 3 clusters to find the most undervalued section which came out to be the red section with around 90 or so players. From there I was able to print out those players names to find the most undervalued ones for my team to attempt to sign.
